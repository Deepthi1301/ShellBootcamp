```python
def solution(S):
    substr_count = 0
    char_positions = {}
    start = 0
    
    for end, char in enumerate(S):
        if char in char_positions and char_positions[char] >= start:
            substr_count += 1
            start = char_positions[char] + 1
        char_positions[char] = end
    
    return substr_count + 1
```
